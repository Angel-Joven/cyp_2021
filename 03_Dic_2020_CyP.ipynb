```
edad = int(input("Introduce tu edad: "))
while edad < 0:
  print("Has introducido una edad negativa, intentalo de nuevo")
  edad = int(input("Introduce tu edad: "))

print("Gracias por participar, puede pasar")
print("Edad del aspirante: " +str (edad))
```

```
edad = int(input("Introduce tu edad: "))
while edad < 5 or edad > 100:
  print("Has introducido una edad negativa, intentalo de nuevo")
  edad = int(input("Introduce tu edad: "))

print("Gracias por participar, puede pasar")
print("Edad del aspirante: " +str (edad))
```
