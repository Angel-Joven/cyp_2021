```
#Problema 4.11

print("Ingrese el total de paises del SUR, CENTRO y NORTEAMERICA: ")
TPS = int(input("Paises del SUR: "))
TPC = int(input("Paises del CENTRO: "))
TPN = int(input("Paises de NORTEAMERICA: "))
TPA = TPS + TPC + TPN + 10

AMERICA = [0 for x in range(TPA)]
SUR = [0 for x in range(TPS)]
CENTRO = [0 for x in range(TPC)]
N_AMERICA = [0 for x in range(TPN)]

if 1 <= TPS and TPS <= 30 and 1 <= TPC and TPC <= 30 and 1 <= TPN and TPN <= 30:
  print("Paises de SUDAMERICA: ")
  for i in range(TPS):
    SUR[i] = input("Nombre del pais: ")
  print("Paises de CENTROAMERICA: ")
  for i in range(TPC):
    CENTRO[i] = input("Nombre del pais: ")
  print("Paises de NORTEAMERICA: ")
  for i in range(TPN):
    N_AMERICA[i] = input("Nombre del pais: ")

  PS = 0
  PC = 0
  PN = 0
  PA = 0

  while PS != TPS and PC != TPC and PN != TPN:
    PA = PA + 1
    if ((SUR[PS] < CENTRO[PC]) and (SUR[PS] < N_AMERICA[PN])):
      AMERICA[PA] = SUR[PS]
      PS = PS + 1
    elif (CENTRO[PC] < N_AMERICA[PN]):
      AMERICA[PA] = CENTRO[PC]
      PC = PC + 1
    else:
      AMERICA[PA] = N_AMERICA[PN]
      PN = PN + 1
  while ((PS != TPS) and (PC != TPC)):
    PA = PA + 1
    if (SUR[PS] < CENTRO[PC]):
      AMERICA[PA] = SUR[PS]
      PS = PS + 1
    else:
      AMERICA[PA] = CENTRO[PC]
      PC = PC + 1
  while ((PS != TPS) and (PN != TPN)):
    PA = PA + 1
    if (SUR[PS] < N_AMERICA[PN]):
      AMERICA[PA] = SUR[PS]
      PS = PS + 1
    else:
      AMERICA[PA] = N_AMERICA[PN]
      PN = PN + 1
  while ((PC != TPC) and (PN != TPN)):
    PA = PA + 1
    if (CENTRO[PC] < N_AMERICA[PN]):
      AMERICA[PA] = CENTRO[PC]
      PC = PC + 1
    else:
      AMERICA[PA] = N_AMERICA[PN]
      PN = PN + 1
  while (PS != TPS):
    PA = PA + 1
    AMERICA[PA] = SUR[PS]
    PS = PS + 1
  while (PC != TPC):
    PA = PA + 1
    AMERICA[PA] = CENTRO[PC]
    PC = PC + 1
  while (PN != TPN):
    PA = PA + 1
    AMERICA[PA] = N_AMERICA[PN]
    PN = PN + 1
  print("Arreglo America")
  for i in range(PA):
    print(f"{i + 1}) {AMERICA[i + 1]}")

else:
  print("Verifique el numero de paises ingresados")
```
